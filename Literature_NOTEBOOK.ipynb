# CAP Theorem
<pre>
C - Consistency: All nodes have a consistent view of the data irrespective of where the client is connected to
A - Availability: Ability to respond to requests from users at all time
P - Partition Tolerance: Ability of system to continue operating even under a <b>network partition<b>
</pre>
IF. a system suffers from network attack, it can prioritize either:<br>
1. Availability but forgoes Consistency 
OR
2. Consistency but forgoes Availability
